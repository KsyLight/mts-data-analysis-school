### Картинки 1 и 2 + лейбл

```python
# Функция, которая готовит и конкатенирует пару изображений
def concat_pair(img_bytes1: bytes, img_bytes2: bytes, resample=Image.LANCZOS) -> np.ndarray:
    im1 = Image.open(io.BytesIO(img_bytes1)).convert('RGB')
    im2 = Image.open(io.BytesIO(img_bytes2)).convert('RGB')

    # Общая высота — минимальная из двух, чтобы не растягивать
    target_h = min(im1.height, im2.height)

    # Пересчитываем ширины, чтобы сохранить пропорции
    new_w1 = int(im1.width * target_h / im1.height)
    new_w2 = int(im2.width * target_h / im2.height)

    # Ресайз с LANCZOS‑интерполяцией (наилучшее сглаживание при уменьшении)
    im1_resized = im1.resize((new_w1, target_h), resample=resample)
    im2_resized = im2.resize((new_w2, target_h), resample=resample)

    # Конкатенация по горизонтали
    return np.concatenate([np.array(im1_resized), np.array(im2_resized)], axis=1)

# Основной блок визуализации
fig, axs = plt.subplots(3, 4, figsize=(12, 8))
for ax in axs.flatten():
    idx = random.randrange(len(ds_tr))
    img1 = ds_tr[idx]["image_1"]
    img2 = ds_tr[idx]["image_2"]
    label = ds_tr[idx]["is_image1_better"]

    concatenated = concat_pair(img1, img2)
    ax.imshow(concatenated)
    ax.axis("off")
    ax.set_title(f"label={label}", fontsize=10)

plt.tight_layout(pad=1.0)
plt.show()
```

## Баланс классов

```python
# Баланс классов для стратифицированного сплита и весов
labels = np.array([ex["is_image1_better"] for ex in ds_tr])
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=SEED)
tr_idx, val_idx = next(sss.split(np.zeros(len(labels)), labels))
ds_train, ds_val = ds_tr.select(tr_idx), ds_tr.select(val_idx)
pos, neg = (labels[tr_idx] == 1).sum(), (labels[tr_idx] == 0).sum()
pos_weight = torch.tensor([neg / pos], device=device)
```

```python
# Класс PairDS для обработки парных изображений (Siamese)
class PairDS(Dataset):
    def __init__(self, hf, tfm, label=True):
        self.ds = hf
        self.tfm = tfm
        self.lbl = label
    def __len__(self):
        return len(self.ds)
    def __getitem__(self, i):
        ex = self.ds[i]
        i1 = Image.open(io.BytesIO(ex["image_1"])).convert("RGB")
        i2 = Image.open(io.BytesIO(ex["image_2"])).convert("RGB")
        if self.lbl:
            y = torch.tensor(ex["is_image1_better"], dtype=torch.float32)
            return self.tfm(i1), self.tfm(i2), y
        return self.tfm(i1), self.tfm(i2), i
```

```python
# Взвешенный сэмплер для устранения дисбаланса классов
def make_sampler():
    w = torch.DoubleTensor([neg / pos if y == 1 else 1.0 for y in labels[tr_idx]])
    return WeightedRandomSampler(w, len(w), replacement=True)
```

### Базовые строительные блоки

```python
class DW(nn.Module):  # Depthwise Conv
    def __init__(self, c, k=3, s=1):
        super().__init__()
        p = k // 2
        self.net = nn.Sequential(
            nn.Conv2d(c, c, k, s, p, groups=c, bias=False),
            nn.Conv2d(c, c, 1, bias=False),
            nn.BatchNorm2d(c), nn.SiLU()
        )
    def forward(self, x):
        return self.net(x)

class SE(nn.Module):  # Squeeze-and-Excitation
    def __init__(self, c, r=16):
        super().__init__()
        self.fc1 = nn.Linear(c, c // r)
        self.fc2 = nn.Linear(c // r, c)
    def forward(self, x):
        w = torch.sigmoid(self.fc2(F.silu(self.fc1(x.mean((2, 3))))))
        return x * w[:, :, None, None]

class TinyEnc(nn.Module):  # Компактный энкодер
    def __init__(self):
        super().__init__()
        self.stem = nn.Sequential(
            nn.Conv2d(3, 64, 7, 2, 3, bias=False),
            nn.BatchNorm2d(64), nn.SiLU(),
            nn.MaxPool2d(3, 2, 1)
        )
        self.b1 = nn.Sequential(DW(64), DW(64), SE(64))
        self.b2 = nn.Sequential(
            nn.Conv2d(64, 128, 3, 2, 1, bias=False),
            nn.BatchNorm2d(128), nn.SiLU(), DW(128), SE(128)
        )
        self.b3 = nn.Sequential(
            nn.Conv2d(128, 256, 3, 2, 1, bias=False),
            nn.BatchNorm2d(256), nn.SiLU(), DW(256), SE(256)
        )
        self.b4 = nn.Sequential(
            nn.Conv2d(256, 512, 3, 2, 1, bias=False),
            nn.BatchNorm2d(512), nn.SiLU(), DW(512), SE(512)
        )
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Linear(512, 512)
    def forward(self, x):
        x = self.stem(x)
        x = self.b1(x); x = self.b2(x); x = self.b3(x); x = self.b4(x)
        return F.normalize(self.fc(self.pool(x).flatten(1)), dim=1)

class SiamCNN(nn.Module):  # Сиамская сеть
    def __init__(self):
        super().__init__()
        self.enc = TinyEnc()
        self.head = nn.Sequential(
            nn.Linear(2048, 512), nn.ReLU(),
            nn.Dropout(0.3), nn.Linear(512, 1)
        )
    def forward(self, x1, x2):
        f1, f2 = self.enc(x1), self.enc(x2)
        v = torch.cat([f1, f2, (f1 - f2).abs(), f1 * f2], 1)
        return self.head(v).squeeze(1)

model_cnn = SiamCNN().to(device)
if torch.cuda.device_count() > 1:
    model_cnn = nn.DataParallel(model_cnn)
```

### Определение функции потерь, оптимизатора, скейлера и scheduler

```python
crit = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
scaler = GradScaler()
opt = AdamW(model_cnn.parameters(), lr=2e-4, weight_decay=1e-4)
sch = get_cosine_schedule_with_warmup(opt, len(train_loader_336) // 5,
                                      len(train_loader_336) * 4)
```

### Цикл обучения

```python
def train_cnn():
    for ep in range(4):
        model_cnn.train(); lg, lb = [], []
        for x1, x2, y in tqdm(train_loader_336, leave=False, desc=f"CNN ep{ep+1}"):
            x1, x2, y = x1.to(device), x2.to(device), y.to(device)
            opt.zero_grad(set_to_none=True)
            with autocast(device_type="cuda"):
                lo = model_cnn(x1, x2)
                loss = crit(lo, y)
            scaler.scale(loss).backward()
            scaler.step(opt)
            scaler.update()
            sch.step()
            lg.append(lo.detach().cpu())
            lb.append(y.cpu())
        auc = roc_auc_score(torch.cat(lb), torch.cat(lg).sigmoid())
        print(f"CNN ep{ep+1} trainAUC {auc:.4f}")

train_cnn()
```

## CLIP ViT-L/14 (336px) — Private Score: 0.63538 | Public Score: 0.66232

### Аугментации

```python
tf_336_train = T.Compose([
    T.Resize(384, interpolation=InterpolationMode.BICUBIC),
    T.RandomResizedCrop(336, scale=(0.8, 1.0), interpolation=InterpolationMode.BICUBIC),
    T.RandomHorizontalFlip(),
    T.ColorJitter(0.08, 0.08, 0.08, 0.02),
    T.RandomPerspective(0.15, p=0.2, interpolation=InterpolationMode.BICUBIC),
    T.ToTensor(),
    T.Normalize([0.48145466,0.4578275,0.40821073],
                [0.26862954,0.26130258,0.27577711])
])
tf_336_eval = T.Compose([
    T.Resize(384, interpolation=InterpolationMode.BICUBIC),
    T.CenterCrop(336),
    T.ToTensor(),
    T.Normalize([0.48145466,0.4578275,0.40821073],
                [0.26862954,0.26130258,0.27577711])
])
```

### Функция ресайза позиционных эмбеддингов

```python
def resize_pe(backbone, img_size=336, patch_size=14):
    pe = backbone.embeddings.position_embedding.weight  # [N, d]
    cls, grid = pe[:1], pe[1:]
    old_grid = int(grid.size(0) ** 0.5)
    new_grid = img_size // patch_size
    grid = grid.reshape(1, old_grid, old_grid, -1).permute(0, 3, 1, 2)
    grid = F.interpolate(grid, size=(new_grid, new_grid), mode="bicubic", align_corners=False)
    grid = grid.permute(0, 2, 3, 1).reshape(new_grid*new_grid, -1)
    new_pe = torch.cat([cls, grid], dim=0)
    backbone.embeddings.position_embedding = nn.Embedding(new_grid*new_grid + 1, pe.size(1))
    backbone.embeddings.position_embedding.weight.data = new_pe
```

### Модель

```python
class SiamCLIP_L(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = CLIPModel.from_pretrained("openai/clip-vit-large-patch14").vision_model
        self.backbone.config.image_size = 336
        self.backbone.embeddings.image_size = 336
        self.backbone.embeddings.patch_size = 14
        # Position embedding адаптация
        if hasattr(self.backbone.embeddings, 'position_ids'):
            del self.backbone.embeddings.position_ids
        resize_pe(self.backbone, img_size=336, patch_size=14)
        self.backbone.requires_grad_(False)
        d = 1024
        self.head = nn.Sequential(
            nn.Linear(d*4, 1024), nn.ReLU(),
            nn.Dropout(0.3), nn.Linear(1024, 1)
        )

    def forward(self, x1, x2):
        # Генерируем position_ids на устройстве батча
        num_pos = self.backbone.embeddings.position_embedding.weight.size(0)
        pos_ids = torch.arange(num_pos, device=x1.device).unsqueeze(0)
        self.backbone.embeddings.position_ids = pos_ids
        f1 = self.backbone(x1).last_hidden_state[:, 0, :]
        f2 = self.backbone(x2).last_hidden_state[:, 0, :]
        v = torch.cat([f1, f2, (f1-f2).abs(), f1*f2], dim=1)
        return self.head(v).squeeze(1)
```

### Инициализация, обучение и основной тренировочный цикл

```python
model_l = SiamCLIP_L().to(device)
if torch.cuda.device_count() > 1:
    model_l = nn.DataParallel(model_l)

crit = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
scaler = torch.amp.GradScaler()
opt = torch.optim.AdamW(model_l.parameters(), lr=2e-4, weight_decay=1e-4)
sched = get_cosine_schedule_with_warmup(opt, len(train_loader_336)//5, len(train_loader_336)*4)

for ep in range(4):
    model_l.train(); lg, lb = [], []
    for x1, x2, y in tqdm(train_loader_336, desc=f"L/14 ep{ep+1}", leave=False):
        x1, x2, y = x1.to(device), x2.to(device), y.to(device)
        opt.zero_grad(set_to_none=True)
        with torch.amp.autocast(device_type="cuda"):
            lo = model_l(x1, x2); loss = crit(lo, y)
        scaler.scale(loss).backward()
        scaler.step(opt); scaler.update(); sched.step()
        lg.append(lo.detach().cpu()); lb.append(y.cpu())
    auc = roc_auc_score(torch.cat(lb), torch.cat(lg).sigmoid())
    print(f"L/14 ep{ep+1} trainAUC {auc:.4f}")
```

### Функция collate для подготовки батча с горизонтальным отражением изображений

```python
def flip_collate(batch):
    x1s = torch.stack([item[0] for item in batch])
    x2s = torch.stack([item[1] for item in batch])
    x1s_flipped = torch.flip(x1s, dims=[3])
    x2s_flipped = torch.flip(x2s, dims=[3])
    idxs = torch.tensor([item[2] for item in batch]) if len(batch[0]) > 2 else torch.arange(len(batch))
    return x1s_flipped, x2s_flipped, idxs
```

```python
flip_loader = DataLoader(
    PairDS(ds_te, tf_336_te, False),
    batch_size=64, shuffle=False,
    num_workers=4, pin_memory=True, persistent_workers=True,
    collate_fn=flip_collate
)
```

## CLIP ViT-B/16 (224px) — Private Score: 0.60761 | Public Score: 0.64694

### Аугментация и DataLoader'ы

```python
tf_224 = T.Compose([
    T.Resize(256, InterpolationMode.BICUBIC),
    T.CenterCrop(224),
    T.ToTensor(),
    T.Normalize([0.48145466, 0.4578275, 0.40821073],
                [0.26862954, 0.26130258, 0.27577711])
])
train_loader_224 = DataLoader(
    PairDS(ds_train, tf_224, True),
    batch_size=96, sampler=make_sampler(),
    num_workers=4, pin_memory=True, persistent_workers=True
)
test_loader_224 = DataLoader(
    PairDS(ds_te, tf_224, False),
    batch_size=96, shuffle=False,
    num_workers=4, pin_memory=True, persistent_workers=True
)
```

### Модель

```python
class SiamCLIP_B(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = CLIPModel.from_pretrained(
            "openai/clip-vit-base-patch16"
        ).vision_model
        self.backbone.requires_grad_(False)
        d = 768
        self.head = nn.Sequential(
            nn.Linear(d*4, 512), nn.ReLU(),
            nn.Dropout(0.3), nn.Linear(512, 1)
        )
    def forward(self, x1, x2):
        f1 = self.backbone(x1).last_hidden_state[:, 0, :]
        f2 = self.backbone(x2).last_hidden_state[:, 0, :]
        v = torch.cat([f1, f2, (f1-f2).abs(), f1*f2], 1)
        return self.head(v).squeeze(1)
```

### Инициализация модели, loss, scaler, optimizer и scheduler

```python
model_b = SiamCLIP_B().to(device)
if torch.cuda.device_count() > 1:
    model_b = nn.DataParallel(model_b)

critB = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
scaler = GradScaler()

opt1 = AdamW(model_b.module.head.parameters(), lr=5e-5, weight_decay=1e-2)
sch1 = get_cosine_schedule_with_warmup(opt1, len(train_loader_224)//10,
                                       len(train_loader_224)*2)
```

### Тренировочный цикл

```python
def run(loader, opt, sch, desc):
    model_b.train(); lg, lb = [], []
    for x1, x2, y in tqdm(loader, leave=False, desc=desc):
        x1, x2, y = x1.to(device), x2.to(device), y.to(device)
        opt.zero_grad(set_to_none=True)
        with autocast(device_type="cuda"):
            lo = model_b(x1, x2)
            loss = critB(lo, y)
        scaler.scale(loss).backward()
        scaler.step(opt)
        scaler.update()
        sch.step()
        lg.append(lo.detach().cpu())
        lb.append(y.cpu())
    auc = roc_auc_score(torch.cat(lb), torch.cat(lg).sigmoid())
    print(f"{desc} AUC {auc:.4f}")

for ep in range(2):
    run(train_loader_224, opt1, sch1, f"B head{ep+1}/2")
```

### Файн-тюн 10-11 блоков ViT

```python
for n, p in model_b.module.backbone.named_parameters():
    if any(f".{i}." in n for i in (10, 11)):
        p.requires_grad = True
opt2 = AdamW(filter(lambda p: p.requires_grad, model_b.parameters()),
             lr=1e-5, weight_decay=1e-2)
sch2 = get_cosine_schedule_with_warmup(opt2, len(train_loader_224)//10,
                                       len(train_loader_224)*2)
for ep in range(2):
    run(train_loader_224, opt2, sch2, f"B fine{ep+1}/2")
```

## META CatBoost (CNN + ViT-B + ViT-L) — Private Score: 0.63399 | Public Score: 0.66343

### OOF предсказания

```python
def infer_prob(model, loader):
    model.eval(); out=[]
    with torch.no_grad(), autocast(device_type="cuda"):
        for x1,x2,_ in loader:
            out.append(torch.sigmoid(model(x1.to(device),x2.to(device))).cpu())
    return torch.cat(out)
```

```python
loader_tr_336 = DataLoader(PairDS(ds_tr,tf_336_te,True),
                         batch_size=64,shuffle=False,
                         num_workers=4,pin_memory=True,persistent_workers=True)
loader_tr_224 = DataLoader(PairDS(ds_tr,tf_224,True),
                         batch_size=64,shuffle=False,
                         num_workers=4,pin_memory=True,persistent_workers=True)

prob_cnn_tr = infer_prob(model_cnn,loader_tr_336)
prob_b_tr = infer_prob(model_b, loader_tr_224)
prob_l_tr = infer_prob(model_l, loader_tr_336)

prob_stack_train=np.vstack([prob_cnn_tr,prob_b_tr,prob_l_tr]).T
```

### Обучение мета-моедил с оптимизацией гиперпараметров через Optuna | Стекинг

```python
def objective(trial):
    bootstrap = trial.suggest_categorical('bootstrap_type', ['Bayesian', 'Bernoulli'])
    params = {
        'iterations'     : trial.suggest_int ('iterations', 300, 1200),
        'depth'          : trial.suggest_int ('depth', 3, 9),
        'learning_rate'  : trial.suggest_float('learning_rate', 1e-3, 3e-2, log=True),
        'l2_leaf_reg'    : trial.suggest_float('l2_leaf_reg', 1e-2, 10,   log=True),
        'bootstrap_type' : bootstrap,
        'subsample'      : trial.suggest_float('subsample', 0.6, 1.0) if bootstrap=='Bernoulli' else None,
        'loss_function'  : 'Logloss',
        'eval_metric'    : 'AUC',
        'task_type'      : 'GPU',
        'devices'        : '0:1',
        'random_seed'    : SEED,
        'early_stopping_rounds': 50,
        'verbose'        : False
    }
    model = CatBoostClassifier(**{k:v for k,v in params.items() if v is not None})
    model.fit(prob_stack_train[tr_idx], labels[tr_idx],
              eval_set=(prob_stack_train[val_idx], labels[val_idx]),
              use_best_model=True)
    pred = model.predict_proba(prob_stack_train[val_idx])[:, 1]
    return roc_auc_score(labels[val_idx], pred)
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=20, timeout=1800)
```

```python
best_params = study.best_params
best_params.update({
    'loss_function':'Logloss',
    'eval_metric'  :'AUC',
    'task_type'    :'GPU',
    'devices'      :'0:1',
    'random_seed'  :SEED,
    'verbose'      :False
})
meta = CatBoostClassifier(**best_params)
meta.fit(prob_stack_train, labels)
```

### Анализ применимости вашего решения в бизнес-процессах: что и как применять, для чего, почему

> Взял 2 лучший мои модели для вывода по пункту

- **CLIP ViT-L/14 (336 px, TTA c flip_collate)**  
  - даёт наивысший индивидуальный AUC среди рассмотренных базовых моделей  
  - может использоваться в задачах модерации и проверки качества, где скорость инференса менее критична, чем точность  
  - горизонтальный флип во время теста улучшает метрику без дополнительного обучения  

- **CatBoost мета-модель (стек CNN + CLIP-B + CLIP-L)**  
  - обучается на out-of-fold вероятностях трёх сетей, комбинируя их выходы  
  - дообучение после обновления базовых моделей обычно занимает несколько минут  
  - пригодна для финального ранжирования изображений в ситуациях, когда требуется чуть выше точность по сравнению с отдельными моделями  

Точность сравнения изображений важна, например, при выборе обложек товаров или фильтрации пользовательских фото. При массовой обработке можно использовать только более быстрые модели (CNN или CLIP-B), а полный стек подключать к меньшей доле запросов, где оправдано дополнительное время инференса.
